```python
import pandas as pd
df = pd.read_csv('311_Service_Requests_2020.csv')
df.head()
```

```python
def f(x):
    return x.split('/')[2].split(' ')[0]
df['year'] = df["Created Date"].apply(f)
```

```python
df_10025 = df[df["Incident Zip"] == 10025.0]
reside_zhixing_excuse = df[df["Incident Zip"] == 10025.0]["Complaint Type"].value_counts()
a = reside_zhixing_excuse.head(10)
```

```python
df_2020 = df_10025[df_10025.year == "2020"]
```

```python
top10_2020 = df_2020.loc[df_2020["Complaint Type"].isin(a.index),:]
```

```python
top10 = top10_2020.groupby("Complaint Type").count().sort_values(by=['Unique Key'],ascending = False)["Unique Key"]
```
